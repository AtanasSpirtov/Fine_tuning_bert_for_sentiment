```python
from transformers import pipeline
import os
import csv
```

```python
class SummarizedArticle:
  def __init__(self, title, publish_date, text):
    self.title = title
    self.publish_date = publish_date
    self.text = text
```

```python
file_path = 'download_news_and_save/raw_news_data.csv'
if (os.path.exists(file_path) == False):
    print(f"The file {file_path} does not exist.")
    %run "download_news_and_save/Get_training_data_and_save.ipynb"
    
```

```python
def load_data_from_csv(csv_file_path):
    articles = []
    
    with open(csv_file_path, 'r', newline='', encoding='utf-8') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        
        for row in csv_reader:
            title = row['Title']
            publish_date = row['Publish Date']
            text = row['Text']
            
            article = SummarizedArticle(title, publish_date, text)
            articles.append(article)

    return articles
```

```python
raw_news = load_data_from_csv(file_path)
```

```python
summarizer_model = pipeline("summarization", model="Falconsai/text_summarization")
summarized_news_file_name = "summarized_news.csv"
fieldnames = ['title', 'publish_date', 'text']

with open(summarized_news_file_name, 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
    csv_writer.writeheader()
    for article in raw_news:
      if(len(article.text) > 10000):
          print(len(article.text))
          continue;
      summarized_text = summarizer_model(article.text, max_length=1000, min_length=20, do_sample=False)
      csv_writer.writerow({
            'title': article.title,
            'publish_date': article.publish_date,
            'text': summarized_text
    })
```
